```python
def predict_func(x,w1,w2,b):
    predict=x[0]*w1+x[1]*w2+b
    return predict
```

```python
def cal_error(predict,x):
    error=x[2]-predict
    return error
```

```python
def step_func(predict):
    if predict>0:
        return 1
    else:
        return 0
```

```python
def update_values(x,w1,w2,b,eta,error):
    w1=w1+eta*error*x[0]
    w2=w2+eta*error*x[1]
    b=b+eta*error
    return w1 , w2 , b
```
